### What Are Derivatives?

The derivative of a function at a point is the rate at which the function is changing at that point â€” essentially, the slope of the tangent line.



This tells you how f(x) changes with respect to x.

```python
from sympy import symbols, diff

x = symbols('x')
f = x**2
derivative = diff(f, x)
print("Output: f'(x):", derivative)
```

### How Derivatives Power Machine Learning

Every time you train a model using gradient descent, you're using derivatives to:

- Understand how each weight affects the loss
- Move in the direction that minimizes the loss



```python
### Practical Example

import numpy as np
import matplotlib.pyplot as plt

# Loss function: L(w) = (w - 3)^2
w = np.linspace(0, 6, 100)
L = (w - 3)**2
dL = 2 * (w - 3)

plt.plot(w, L, label='Loss')
plt.plot(w, dL, label='Gradient')
plt.axvline(3, linestyle='--', color='gray', label='Minima at w=3')
plt.legend()
plt.xlabel('w')
plt.title('Loss and its Derivative')
plt.grid(True)
plt.show()
```

What this shows:

- ðŸ“‰ Loss curve: Parabola with minimum at w = 3
- ðŸŸ§ Derivative curve: Linear; zero when w = 3, negative to the left, positive to the right
- ðŸŽ¯ Gradient Descent uses the derivative to decide whether to move left or right, and how far

### Interview Focus

1. What is a derivative in the context of gradient descent?

> A derivative in gradient descent represents the rate of change of the loss function with respect to a model parameter. It tells us how the loss will change if we slightly increase or decrease that parameter â€” guiding us in choosing the direction for optimization.

2. Why is the derivative 0 at the minimum of a loss function?

> At the minimum of a loss function, the slope (i.e., the derivative) is zero because the function is neither increasing nor decreasing at that point. This is a critical point, and in convex functions, it represents the global minimum.

3. How does a learning rate interact with gradients during optimization?

> The learning rate determines the size of the step taken in the direction of the negative gradient during optimization. A small learning rate leads to slow convergence, while a large one may cause overshooting or instability.
